# Data Privacy and Security:
1. Data Privacy: Refers to the appropriate handling of personal data, ensuring that individuals have control over how their information is collected, used, and shared.

2. Data Security: Focuses on safeguarding data against unauthorized access, disclosure, or alteration to ensure its confidentiality, integrity, and availability.

**Data Privacy and Security Techniques:**
  1. Anonymization: It is the process of irreversibly transforming data in such a way that it can no longer be associated with an individual. The goal is to render the data anonymous, meaning that even if someone gains access to the data, they cannot identify the individuals to whom the data belongs.
  2. Pseudonymization: It involves replacing direct identifiers in data with artificial identifiers, known as pseudonyms. Unlike anonymization, pseudonymization allows for the possibility of re-identification but only by authorized parties who possess the corresponding pseudonymization key.

Data anonymization in Python techniques:
1. Randomization:
  
  Randomization involves replacing sensitive data with random values. For example, a person's name might be replaced with a randomly generated string of characters.

```python
import pandas as pd
import random
import numpy as np


# Load the dataset
df = pd.read_csv('employee_hr.csv')
df
```

```python
import string


# define function that operates on entire dataframe
def randomize_values(df):
    for column in df.columns:
        if df[column].dtype == 'O': # check if column has object dtype
            df[column] = [''.join(random.choices(string.ascii_letters + string.digits, k=10))
            for _ in range(len(df))] # generate a list of random strings
    return df
```

```python
# add a new column with unique integer-only IDs
df['UniqueID'] = list(range(1001, 1101))
df
```

```python
# apply function to dataframe
df_rand1 = randomize_values(df)

df_rand1
```

 2. Generalization:
  
  Generalization involves replacing specific values with more general or less precise values.
  It involves transforming or categorizing data into broader categories or ranges.
  For example, instead of storing exact ages, you might generalize ages into age groups (e.g., 20-30, 30-40, etc.).

```python
df_age_salary = pd.read_csv('employee_hr.csv')

# Anonymize the ages by grouping them into age ranges
bins = [20, 30, 40, 50, 60]
labels = ['20-29', '30-39', '40-49', '50-59']
df_age_salary['Age'] = pd.cut(df_age_salary['Age'], bins=bins, labels=labels)



# Anonymize the salaries by grouping them into ranges
bins = [0, 30_000, 50_000, 80_000, 100_000, 200_000]
labels = ['0-30K', '31-50K', '51-80K', '81-100K', '100K+']
df_age_salary['Salary'] = pd.cut(df_age_salary['Salary'], bins=bins, labels=labels)




df_age_salary
```

4. Perturbation:

  Perturbation involves adding random noise to sensitive data to make it harder to recognize. For example, a person's salary might be increased or decreased by a small amount to protect their privacy. The amount of noise added can be adjusted to balance privacy with data utility.

```python
import numpy as np

dfp = pd.read_csv('employee_hr.csv')


def add_noise(dfp, column, std = None):   #std=standaard deviation
    if std == None:
        std = dfp[column].std()

    withNoise = dfp[column].add(np.random.normal(0, std, dfp.shape[0]))
    copy = dfp.copy()
    copy[column] = withNoise
    return copy

perturbation = add_noise(dfp, 'Salary', std=100)
perturbation
```

5. Aggregation:

  Aggregation involves combining data from multiple individuals to create a group-level view of the data. For example, instead of storing data for each individual separately, data might be aggregated into ranges or groups.

```python
import pandas as pd

# Sample transactions data
data = {
    'Date': ['2022-01-01', '2022-01-15', '2022-02-05', '2022-02-20', '2022-03-10'],
    'Amount': [100, 150, 200, 120, 180]
}

# Create DataFrame
df = pd.DataFrame(data)

# Convert 'Date' column to datetime format
df['Date'] = pd.to_datetime(df['Date'])

# Extract month from 'Date' column
df['Month'] = df['Date'].dt.strftime('%Y-%m')  #string format time

# Aggregate data by month
aggregated_data = df.groupby('Month')['Amount'].sum().reset_index()

print("Aggregated Data:")
print(aggregated_data)
```

1. Scrambling:

  Scrambling involves rearranging or shuffling the characters in a string.

```python
df = pd.read_csv("medical_records.csv")

# Define a function to scramble data in a specific column
def scramble_data(data):
    data_list = list(data)
    random.shuffle(data_list)
    return ''.join(data_list)

# Scramble the 'Medical Condition' column
df['Scrambled_medical_condition'] = df['Medical Condition'].apply(scramble_data)

# Display the result
print(df[['Medical Condition', 'Scrambled_medical_condition']].head())
```

2. Tokenization:

  Tokenization involves splitting a string into individual tokens (words, phrases, symbols, etc.).

```python
import nltk  #Natural Language Toolkit (NLTK)
from nltk.tokenize import word_tokenize
df_token= pd.read_csv('medical_records.csv')


# Download punkt tokenizer
nltk.download('punkt') #pre-trained unsupervised machine learning model provided by NLTK for tokenizing text into sentences.


# Define a function to tokenize data in a specific column
def tokenize_data(data):
   return word_tokenize(data)

# Tokenize the 'Medical Condition' column
df_token['Tokenized_Treatment'] = df_token['Treatment'].apply(tokenize_data)

df_token
```

3. Hashing:

  Hashing involves converting data into a fixed-size string of bytes using a hash function.

```python
import hashlib

# Hash the 'Treatment' column using a lambda function
df['Hashed_Treatment'] = df['Treatment'].apply(lambda x: hashlib.sha256(x.encode()).hexdigest()) # Secure Hash Algorithm 256-bit

# Display the result
df
```

4. Encryption:

  Encryption involves transforming data into a ciphertext using an encryption algorithm and a secret key.

```python
from cryptography.fernet import Fernet

df_hash = pd.read_csv('medical_records.csv')

# Generate a key
key = Fernet.generate_key()

# Initialize cipher
cipher = Fernet(key)

# Encrypt the 'Date of Treatment' column using a lambda function
df_hash['Encrypted_Date_of_Treatment'] = df_hash['Date of Treatment'].apply(lambda x: cipher.encrypt(x.encode()).decode())
df_hash
```

5. Format-Preserving Encryption (FPE):

  Format-preserving encryption is a type of encryption that preserves the format of the original data
